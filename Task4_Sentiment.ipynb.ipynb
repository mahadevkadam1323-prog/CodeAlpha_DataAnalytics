```python
from textblob import TextBlob
import pandas as pd
```

```python

reviews = [
    "I love this course!",
    "The classes are boring.",
    "I feel okay about my performance.",
    "Amazing experience with the teachers!",
    "I hate doing homework.",
    "The lectures are very helpful.",
    "I am not satisfied with my grades.",
    "I enjoy studying with my friends.",
    "The course is too difficult.",
    "I am happy with my progress."
]


df_reviews = pd.DataFrame(reviews, columns=['Review'])
df_reviews.head()
```

```python
# Function to classify sentiment
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return "Positive"
    elif analysis.sentiment.polarity < 0:
        return "Negative"
    else:
        return "Neutral"

# Apply to all reviews
df_reviews['Sentiment'] = df_reviews['Review'].apply(get_sentiment)
df_reviews
```

```python
import matplotlib.pyplot as plt

# Count of each sentiment
sentiment_counts = df_reviews['Sentiment'].value_counts()

# Bar chart
sentiment_counts.plot(kind='bar', color=['green','red','blue'])
plt.title("Sentiment Analysis of Reviews")
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")
plt.show()
```

1. Positive reviews: Shows students who are happy with the course.
2. Negative reviews: Shows students who are unhappy or facing difficulties.
3. Neutral reviews: Shows students with mixed feelings.
4. The overall sentiment helps understand how students feel about their learning experience.
