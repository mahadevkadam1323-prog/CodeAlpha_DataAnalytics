```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/StudentsPerformance.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Average total score by grade
avg_scores = df.groupby('grade')['total_score'].mean()

# Plot
avg_scores.plot(kind='bar', color='skyblue')
plt.title("Average Total Score by Grade")
plt.ylabel("Average Score")
plt.xlabel("Grade")
plt.show()
```

```python
plt.hist(df['weekly_self_study_hours'], bins=10, color='lightgreen')
plt.title("Distribution of Weekly Self Study Hours")
plt.xlabel("Hours per Week")
plt.ylabel("Number of Students")
plt.show()
```

```python
sns.boxplot(x='attendance_percentage', y='total_score', data=df)
plt.title("Attendance Percentage vs Total Score")
plt.xlabel("Attendance Percentage")
plt.ylabel("Total Score")
plt.show()
```

```python
plt.scatter(df['weekly_self_study_hours'], df['total_score'], color='orange')
plt.title("Weekly Self Study Hours vs Total Score")
plt.xlabel("Weekly Self Study Hours")
plt.ylabel("Total Score")
plt.show()
```
